Here I changed the DataType of Date Column

It looks through the DataFrame (campaigns) and picks out columns with the data type object.
It searches for any NaN (Not a Number) or None values within those specific columns. These represent empty cells or missing information.
It swaps those empty spots with the string "Unknown" and saves the changes back into the original DataFrame.

```python
text_cols = campaigns.select_dtypes(include='object').columns
campaigns[text_cols] = campaigns[text_cols].fillna("Unknown")
```

```python
num_cols = campaigns.select_dtypes(include=['int64','float64']).columns
campaigns[num_cols] = campaigns[num_cols].fillna(0)
```

Rechecked the null values

```python
campaigns.isnull().sum()
```

Here I Filled the Date Column Missing Values with Forward Fill

```python
campaigns['Date'] = campaigns['Date'].ffill()
```

CTR

```python
campaigns.loc[:, 'CTR'] = campaigns['Clicks (all)'] / campaigns['Impressions']
```

```python
campaigns[['Clicks (all)', 'Impressions', 'CTR']].head()
```

CPM

```python
campaigns.loc[:, 'CPM'] = (campaigns['Amount Spent (INR)'] / campaigns['Impressions']) * 1000
campaigns.loc[campaigns['Impressions'] == 0, 'CPM'] = 0
```

```python
campaigns[['CTR','CPM']].head()
```

ROI

```python
campaigns.loc[:, 'ROI'] = (
    campaigns['Purchases Conversion Value (INR)'] - campaigns['Amount Spent (INR)']
) / campaigns['Amount Spent (INR)']

campaigns.loc[campaigns['Amount Spent (INR)'] == 0, 'ROI'] = 0
```

CPC

```python
campaigns.loc[:, 'CPC'] = campaigns['Amount Spent (INR)'] / campaigns['Clicks (all)']
campaigns.loc[campaigns['Clicks (all)'] == 0, 'CPC'] = 0
```

```python
campaigns[['CTR','CPC','CPM','ROI']].head()
```

```python
text_cols = campaigns.select_dtypes(include='object').columns
text_cols
```

```python
campaigns[text_cols] = campaigns[text_cols].apply(lambda x: x.str.strip().str.upper())
```

```python
campaigns[text_cols] = campaigns[text_cols].fillna("UNKNOWN")
```
