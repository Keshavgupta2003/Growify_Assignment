```python
# Convert timestamps to datetime objects
Raw['Transaction Timestamp'] = pd.to_datetime(Raw['Transaction Timestamp'], errors='coerce')
Raw['Order Created At'] = pd.to_datetime(Raw['Order Created At'], errors='coerce')

# Fill missing 'Date' values using the date part of 'Transaction Timestamp'
Raw['Date'] = pd.to_datetime(Raw['Date'], errors='coerce')
Raw['Date'] = Raw['Date'].fillna(Raw['Transaction Timestamp'].dt.date)
Raw['Date'] = Raw['Date'].fillna(Raw['Order Created At'].dt.date)

# Ensure 'Date' is in a standard YYYY-MM-DD format
Raw['Date'] = pd.to_datetime(Raw['Date']).dt.date
```

```python
# List of categorical columns to clean
import numpy as np
cat_cols = ['Sales Channel', 'Billing Country', 'Billing City', 'Product Type', 'Shipping Country']

for col in cat_cols:
    # Replace literal 'NAN' or 'nan' strings with actual Nulls
    Raw[col] = Raw[col].replace(['NAN', 'nan', 'NaN'], np.nan)
    
    # Standardize casing (Title Case for consistency)
    Raw[col] = Raw[col].astype(str).str.title().str.strip()
    
    # Re-mask the 'Nan' string back to a real null if it was created by the .str.title()
    Raw[col] = Raw[col].replace('Nan', np.nan)
```

```python
# Fix Order ID: Convert from scientific notation to string
Raw['Order ID'] = Raw['Order ID'].apply(lambda x: '{:.0f}'.format(x) if pd.notnull(x) else x)

# Ensure numeric columns are floats
numeric_cols = ['Gross Sales (INR)', 'Net Sales (INR)', 'Discounts (INR)', 'Returns (INR)']
for col in numeric_cols:
    Raw[col] = pd.to_numeric(Raw[col], errors='coerce').fillna(0.0)

# Recalculate Net Sales to ensure accuracy
# Net Sales = Gross Sales - Discounts - Returns
Raw['Net Sales (INR)'] = Raw['Gross Sales (INR)'] - Raw['Discounts (INR)'] - abs(Raw['Returns (INR)'])

# Calculate Return Rate if missing
Raw['Return Rate'] = Raw.apply(lambda row: (row['Items Returned'] / row['Items Sold']) 
                             if row['Items Sold'] > 0 else 0, axis=1)
```

```python
def generate_quality_report(Raw):
    report = {
        "Total Rows": len(Raw),
        "Missing Dates Fixed": Raw['Date'].notnull().sum(),
        "Unique Sales Channels": Raw['Sales Channel'].nunique(),
        "Total Net Sales (INR)": Raw['Net Sales (INR)'].sum()
    }
    print("--- DATA QUALITY REPORT ---")
    for key, value in report.items():
        print(f"{key}: {value}")

generate_quality_report(Raw)

# Export the cleaned data for Task 2 (SQL)
Raw.to_csv('Cleaned_Shopify_Sales.csv', index=False)
```
